## Part 1: Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import numpy as np
from tensorflow.keras.models import Model
from tensorflow.keras import layers
from tensorflow.keras.layers import Input, Dense, concatenate

#  Import and read the attrition data
attrition_df = pd.read_csv('https://static.bc-edx.com/ai/ail-v-1-0/m19/lms/datasets/attrition.csv')
attrition_df.head()
```

```python
# Create y_df with the Attrition and Department columns
y_df = attrition_df[['Attrition', 'Department']]

y_df.head()
```

```python
# Create a list of at least 10 column names to use as X data
X_column_list = ['Age', 'BusinessTravel', 'DistanceFromHome',
                 'EnvironmentSatisfaction', 'HourlyRate',
                 'JobInvolvement', 'JobLevel', 'JobRole', 'JobSatisfaction',
                 'NumCompaniesWorked', 'OverTime', 'PercentSalaryHike',
                 'PerformanceRating', 'RelationshipSatisfaction', 'StockOptionLevel',
                 'TotalWorkingYears', 'TrainingTimesLastYear', 'WorkLifeBalance',
                 'YearsAtCompany', 'YearsInCurrentRole', 'YearsSinceLastPromotion',
                 'YearsWithCurrManager']


# Create X_df using your selected columns
X_df = attrition_df[X_column_list]

# Show the data types for X_df
X_df.dtypes
```

```python
# Split the data into training and testing sets
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_df, y_df)

```

```python
# Check for null values
null_counts = X_df.isnull().sum()
print(null_counts)
```

```python
# Create lists for OneHotEncoder and LabelEncoder

enc_columns = ['BusinessTravel', 'OverTime']
LE_columns = ['JobRole']
```

```python
# Create a OneHotEncoder for BusinessTravel and Overtime columns
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(drop='if_binary', sparse=False)

# encode the columns included in the enc_columns list. Do this for both train and test sets.
X_encoded_array_train = encoder.fit_transform(X_train[enc_columns])
X_encoded_array_test = encoder.fit_transform(X_test[enc_columns])

print(X_encoded_array_train[0:5])
print(X_encoded_array_test[0:5])
```

```python
# use get feature names to get list of columns
encoded_columns = encoder.get_feature_names_out(enc_columns)

# convert encoded arrays into dataframes
X_enc_train = pd.DataFrame(X_encoded_array_train, columns=encoded_columns)
X_enc_test = pd.DataFrame(X_encoded_array_test, columns=encoded_columns)

X_enc_test
```

```python
# Create Label Encoder for JobRole column. Using Label encoder because there are 9 different possible responses and want to avoid adding 9 additional columns.

from sklearn.preprocessing import LabelEncoder

LE = LabelEncoder()

# Create copy of X_df to use Label Encoder. Will then concatenate that LE dataframe with the OneHotEncoder df. Will need to remember to drop the original un-encoded columns

X_train_label_encoded = X_train.copy()
X_test_label_encoded = X_test.copy()

for column in LE_columns:
  X_train_label_encoded[column] = LE.fit_transform(X_train_label_encoded[column])
  X_test_label_encoded[column] = LE.fit_transform(X_test_label_encoded[column])

X_train_label_encoded.reset_index(drop=True, inplace=True)
X_test_label_encoded.reset_index(drop=True, inplace=True)

X_test_label_encoded


```

```python
# Merge OneHotEncoded dataframe to label encoded dataframe and drop original unencoded columns
# Used merge because putting together dataframes with different column names.

X_train_final = pd.merge(X_train_label_encoded, X_enc_train, left_index=True, right_index=True, how='inner')
X_test_final = pd.merge(X_test_label_encoded, X_enc_test, left_index=True, right_index=True, how='inner')


X_train_final.reset_index(drop=True, inplace=True)
X_test_final.reset_index(drop=True, inplace=True)

X_train_final.drop(['BusinessTravel', 'OverTime'], axis=1, inplace=True)
X_test_final.drop(['BusinessTravel', 'OverTime'], axis=1, inplace=True)

X_test_final
```

```python
X_train_final.columns
```

```python
# Create a StandardScaler
X_scaler = StandardScaler()

# Fit the StandardScaler to the training data
X_scaler.fit(X_train_final)

# Scale the training and testing data
X_train_scaled = X_scaler.transform(X_train_final)
X_test_scaled = X_scaler.transform(X_test_final)
```

```python
# Create a OneHotEncoder for the Department column
y_encoder_dept = OneHotEncoder(sparse=False)

# Fit the encoder to the training data
# Create two new variables by applying the encoder
# to the training and testing data

y_encoded_dept_train = y_encoder_dept.fit_transform(y_train[['Department']])
y_encoded_dept_test = y_encoder_dept.fit_transform(y_test[['Department']])

print(y_encoded_dept_train[0:5])
print(y_encoded_dept_test[0:5])
```

```python
# Create a OneHotEncoder for the Attrition column
y_encoder_attr = OneHotEncoder(sparse=False)

# Fit the encoder to the training data
# Create two new variables by applying the encoder
# to the training and testing data

y_encoded_attr_train = y_encoder_dept.fit_transform(y_train[['Attrition']])
y_encoded_attr_test = y_encoder_dept.fit_transform(y_test[['Attrition']])

print(y_encoded_attr_train[0:5])
print(y_encoded_attr_test[0:5])
```

## Create, Compile, and Train the Model

```python
# Find the number of columns in the X training data
input_nodes = len(X_train_final.columns)

# Create the input layer
input_layer = layers.Input(shape=(input_nodes,), name='input_layer')

# Create at least two shared layers. Started with 24 neurons, because that is the number of inputs. Then tried again with 48.
shared_layer1 = layers.Dense(48, activation='relu', name='shared_layer1')(input_layer)
shared_layer2 = layers.Dense(24, activation= 'relu', name='shared_layer2')(shared_layer1)
```

```python
# Create a branch for Department
# with a hidden layer and an output layer


# Create the hidden layer
dept_hidden = Dense(6, activation='relu', name='dept_hidden')(shared_layer2)

# Create the output layer. Using softmax because there are 3 possible departments and want to know the probability in relation to each other.
dept_output = Dense(3, activation='softmax', name='dept_output')(dept_hidden)
```

```python
# Create a branch for Attrition
# with a hidden layer and an output layer

# Create the hidden layer
attr_hidden = Dense(6, activation='relu', name='attr_hidden')(shared_layer2)

# Create the output layer
attr_output = Dense(2, activation='sigmoid', name='attr_output')(attr_hidden)
```

```python
# Create the model
model = Model(inputs=input_layer, outputs=[dept_output, attr_output])

# Compile the model
model.compile(optimizer='adam',
              loss={'dept_output': 'categorical_crossentropy',
                    'attr_output': 'binary_crossentropy'},
              metrics={'dept_output': 'accuracy',
                       'attr_output': 'accuracy'})

# Summarize the model
model.summary()
```

```python
# Train the model
model.fit(
    X_train_final,
    {'dept_output': y_encoded_dept_train, 'attr_output': y_encoded_attr_train},
    epochs = 20,
    batch_size = 32
)

```

```python
# Evaluate the model with the testing data
results = model.evaluate(np.array(X_test_final), {
    'dept_output': y_encoded_dept_test,
    'attr_output': y_encoded_attr_test
})
```

```python
# Print the accuracy for both department and attrition
pred_categories = ['Attrition', 'Department']
for i, cat in enumerate(pred_categories):
    print(f"{cat} accuracy: {results[i+3]}")
```
